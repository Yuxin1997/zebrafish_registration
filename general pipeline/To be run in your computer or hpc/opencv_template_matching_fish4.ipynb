```python
# fish1 120um one

img = tifffile.imread('fish4.tif')
img2 = img.copy()
# img = cv.imread('AVG_motion_corrected_f1.tif',0)
# img2 = img.copy()
template = cv.imread('2d_120/AVG_motion_corrected_f4_120.tif',0)
w, h = template.shape[::-1]
method = cv.TM_CCOEFF_NORMED

# top_left = np.array(shape=(img2.shape[0],2))
corr_score_120 = []

for i in range(0,img2.shape[0]):
    img = img2[i].copy()
    res = cv.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
    # top_left = max_loc
    corr_score_120.append(max_val)

```

```python
# fish1 60um one

img = tifffile.imread('fish4.tif')
img2 = img.copy()
# img = cv.imread('AVG_motion_corrected_f1.tif',0)
# img2 = img.copy()
template = cv.imread('2d_60/AVG_motion_corrected_f4_60.tif',0)
w, h = template.shape[::-1]
method = cv.TM_CCOEFF_NORMED

# top_left = np.array(shape=(img2.shape[0],2))
corr_score_60 = []

for i in range(0,img2.shape[0]):
    img = img2[i].copy()
    res = cv.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
    # top_left = max_loc
    corr_score_60.append(max_val)
```

```python
dist120, = plt.plot(corr_score_120)
dist100, = plt.plot(corr_score_100)
dist80, = plt.plot(corr_score_80)
dist60, = plt.plot(corr_score_60)
plt.text(np.argmax(corr_score_120),max(corr_score_120)+0.01,str(np.argmax(corr_score_120)))
plt.text(np.argmax(corr_score_100),max(corr_score_100)+0.01,str(np.argmax(corr_score_100)))
plt.text(np.argmax(corr_score_80),max(corr_score_80)+0.01,str(np.argmax(corr_score_80)))
plt.text(np.argmax(corr_score_60),max(corr_score_60)+0.01,str(np.argmax(corr_score_60)))
plt.ylim([0,0.65])
plt.xlabel('Slice No.')
plt.ylabel('Correlation Score')

plt.legend((dist120,dist100,dist80,dist60),('120um plane','100um plane','80um plane','60um plane'))
plt.title('1130F4 template matching')
plt.show()
```
